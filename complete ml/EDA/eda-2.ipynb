```python
#feature engineering
df['date_of_journey']=df['date'].str.split('-').str[0]
df['month_of_journey']=df['date'].str.split('-').str[1]
df['year_of_journey']=df['date'].str.split('-').str[2]
```

```python
df['date_of_journey']=df['date_of_journey'].astype(int)
df['month_of_journey']=df['month_of_journey'].astype(int)
df['year_of_journey']=df['year_of_journey'].astype(int)
```

```python
df['arr_time'] = df['arr_time'].astype('object')   # Extracts the hour part (e.g., "21" from "21:45")
# Extracts the minute part (e.g., "45" from "21:45")
df['arr_hr']=df['arr_time'].str.split(':').str[0]
df['arr_min']=df['arr_time'].str.split(':').str[1]
```
