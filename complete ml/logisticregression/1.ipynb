```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
##create the datsets 
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
#model training 
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
```

```python
logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm= confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

Hyperparameter Tuning And Cross validation


```python
model=LogisticRegression()
penalty=['11','12','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','ibfgs','libinear','sag','saga']
```

```python
param=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedGroupKFold
cv=StratifiedGroupKFold
```

```python
##grid search cv
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=param,scoring='accuracy_score',cv=cv,n_jobs=-1)
```

```python
# grid.fit(X_train,y_train)
grid.best_params_
```

```python
model= LogisticRegression()
random= RandomizedSearchCV(estimator=model,param_distributions=param,cv=5,scoring='accuracy_score')
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=3,n_informative=3,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm= confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
